# RECOMMENDATION SYSTEM USING MATRIX FACTORISATION(SVD)

Singular value decomposition (SVD) is a collaborative filtering method for movie recommendation.

```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python
#import movies dataset
MOVIES=pd.read_csv("movies.csv")
```

```python
MOVIES
```

```python
#Import Ratings dataset
RATINGS=pd.read_csv("ratings.csv")
```

```python
RATINGS
```

```python
#MERGING MOVIES and RATINGS
MOVIES_RATINGS=pd.merge(RATINGS,MOVIES,how='outer',on='movieId')
```

```python
MOVIES_RATINGS
```

```python
#Create a cross_Table for with user_id,movie_id having values of ratings
Cross_Table=RATINGS.pivot(index = 'userId', columns ='movieId', values = 'rating').fillna(0)
```

```python
Cross_Table
```

```python
#Convert it to matrix
Cross_Table_matrix=np.asmatrix(Cross_Table)
```

```python
Cross_Table_matrix
```

```python
#Take the mean of all ratings for each user by taking mean of cross table alomg the coloumn
user_mean_ratings=np.mean(Cross_Table_matrix,axis=1)
```

```python
user_mean_ratings
```

```python
#Demean or normalise the data
Demeaned_Data=Cross_Table_matrix-user_mean_ratings
```

```python
Demeaned_Data
```

```python
#Matrix Fcatorisation using singular value decomposition
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(Demeaned_Data, k = 50)
```

```python
U
```

```python
sigma
```

```python
#Sigma returned should be converted to daigonal matrix
sigma=np.diag(sigma)
```

```python
sigma
```

```python
Vt
```

```python
#Vt.shape
```

```python
# Predict the movie rating for every user
predicted_ratings=np.dot(np.dot(U, sigma), Vt) + user_mean_ratings.reshape(-1, 1)
```

```python
predicted_ratings
```

```python
#Construct preditced matrices
Predicted_Matrix= pd.DataFrame(data=predicted_ratings,columns=Cross_Table.columns)
```

```python
Predicted_Matrix
```

```python
#Creating a function that will return recommended Movies and User Data
def recommend_movies(preds_df, userID, movies_df, original_ratings_df, num_of_recomd=5):
    # Get and sort the user's predictions
    user_row_number = userID - 1 # As UserID starts at 1, not 0
    #Get the information from predicted Matrix of that paricular User
    sorted_user_predictions = preds_df.iloc[user_row_number].sort_values(ascending=False)
    
    # Get the user's data from ratings datset by comparing the userid and then merge in the movies dataframe on basis of movieID.
    user_data = original_ratings_df[original_ratings_df.userId == (userID)]
    user_full_Data= (user_data.merge(movies_df, how = 'left', left_on = 'movieId', right_on = 'movieId').
                     sort_values(['rating'], ascending=False))
    
    
    #Printing the prdicted Movies
    print(f"User {userID} has already rated {user_full_Data.shape[0]} movies")
    print(f"Recommending the top {num_of_recomd} movies that you may like")
    
    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    recommendations = (movies_df[movies_df['movieId'].isin(user_full_Data['movieId'])].
            merge(pd.DataFrame(sorted_user_predictions).reset_index(), how="outer").
        sort_values(user_row_number,ascending=False).   #sorting values on the basis of ratings
                      iloc[:num_of_recomd, :])
    recommendations=recommendations.rename(columns={user_row_number:"Top Ratings"})
    return user_full_Data, recommendations

already_rated, recommd = recommend_movies(Predicted_Matrix, 837,MOVIES ,RATINGS, 10)
```

```python
USER_RATINGS=already_rated["rating"].iloc[:10]# as we have taken 10 number of recommendation
USER_RATINGS=USER_RATINGS.array
PREDICTED_RATINGS=recommd["Top Ratings"]
PREDICTED_RATINGS=PREDICTED_RATINGS.array
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(USER_RATINGS,PREDICTED_RATINGS))
```
